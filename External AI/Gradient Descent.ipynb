```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function and derivative
def f(x):
    return x**2

def df(x):
    return 2*x
```

```python
# parameters
x = 10 # starting point
alpha = 0.1 # learning rate
history = [x]
```

```python
# Gradient Descent loop
for _ in range(20):
    grad = df(x)
    x = x - alpha*grad
    history.append(x)
```

```python
# plot
X = np.linspace(-10, 10, 100)
y = f(X)
plt.plot(X, y, label='f(x)=x^2')
plt.scatter(history, [f(i) for i in history], c='r')
plt.plot(history, [f(i) for i in history], c='r', ls='--', label='Gradient Descent Path')
plt.legend()
plt.show()
```

## Gradient Descent for Linear Regression

```python
x = np.array(range(1, 6))
y = np.array(range(2, 11, 2))
```

```python
# parameters
w, b = 0.0, 0.0 # initial weight and bias
alpha = 0.01
epochs = 1000
n = len(x)
```

```python
# history for visualization
history_w, history_b, history_loss = [], [], []
```

```python
# Gradient Descent Loop
for _ in range(epochs):
    y_pred = w*x + b
    error = y - y_pred

    # gradients 
    dw = (-2/n) * np.sum(x*error)
    db = (-2/n) * np.sum(error)

    # update parameters
    w = w - alpha*dw
    b = b - alpha*db

    # store history
    history_w.append(w)
    history_b.append(b)
    history_loss.append(np.mean(error*2))

print(f'Trained weight: {w:.2f}, bias: {b:.2f}')
```

```python
# plot loss over iterations
plt.plot(history_loss)
plt.title('Loss decrease over iterations')
plt.xlabel('Iteration')
plt.ylabel('MSE Loss')
plt.show()
```

```python
# plot final regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, w*x + b, color='red', label='Fitted Line')
plt.legend()
plt.show()
```
