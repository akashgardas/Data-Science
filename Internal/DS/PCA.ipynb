## What is it?
- Preserves as much of the original variance as possible.
- Compressed features are called **Principle Components**.

## How it works?
- It finds the new axes for the data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt 

import seaborn as sns
```

```python
X, y = load_iris(return_X_y=True)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
print("Comparison Before and After PCA")
print(f'Original Shape: {X_scaled.shape}')
print(f'Reduced Shape: {X_pca.shape}')
```

```python
total_variance = sum(pca.explained_variance_ratio_)
total_variance
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], cmap='viridis', c=y)
plt.show()
```
