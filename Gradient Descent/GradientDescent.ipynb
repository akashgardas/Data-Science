# Gradient Descent
- slope
- small step towards the slope
- repeat until no further slope

**Loss Function**: How wrong the model is. Gradient Descent is used to minimize that loss by updating model parameters step-by-step.



---
### Example:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# function and derivative
def f(x):
    return x**2

def df(x):
    return 2*x
```

```python
# parameters
x = 10 # starting point
alpha = 0.1 # learning rate
history = [x]
```

```python
# Gradient desscent loop
for _ in range(20):
    grad = df(x)
    x = x - alpha * grad
    history.append(x)
```

```python
# Plot
X = np.linspace(-10, 10, 100)
Y = f(X)
```

```python
plt.plot(X, Y, label='f(x)=x^2')
plt.scatter(history, [f(i) for i in history], color='red')
plt.plot(history, [f(i) for i in history], 'r--', label='Gradient Descent Path')
plt.legend()
plt.title('Gradient Descent on f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

## Gradients for Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample data
x = np.array(range(1, 6))
y = np.array(range(2, 11, 2))
x, y
```

```python
# Parameters
w, b = 0.0, 0.0 # initial weight and bias
alpha = 0.01 # learning rate
epochs = 1000
n = len(x)
```

```python
# Store history for visualization
history_w, history_b, history_loss = [], [], []
```

```python
# Gradient Descent Loop
for _ in range(epochs):
    y_pred = w*x + b
    error = y - y_pred

    # gradients
    dw = (-2/n) * np.sum(x*error)
    db = (-2/n) * np.sum(error)

    # update parameters
    w -= alpha * dw
    b -= alpha * db

    # store history
    history_w.append(w)
    history_b.append(b)
    history_loss.append(np.mean(error*2))

print(f'Trained weight: {w:.2f}, bias: {b:.2f}')
```

```python
# plot loss over iterations
plt.plot(history_loss)
plt.title('Loss decrease over iterations')
plt.xlabel('Iteration')
plt.ylabel('MSE Loss')
plt.show()
```

```python
# plot final regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, w*x + b, color='red', label='Fitted Line')
plt.legend()
plt.show()
```
